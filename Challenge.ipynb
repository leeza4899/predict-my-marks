```python
!pip install pandas
!pip install seaborn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Data Prep

```python
X = pd.read_csv("./Train/Linear_X_Train.csv")
y = pd.read_csv("./Train/Linear_Y_Train.csv")


#visualization
plt.style.use("seaborn")
plt.scatter(X,y)
plt.xlabel("Hardwork")
plt.ylabel("Performance")
plt.title("Marks vs Performance Graph")
plt.show()
```

# Linear Reg

```python
!pip install scikit-learn
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize = True)
lr.fit(X,y)
```

```python
lr.predict([[10]])
```

```python
plt.scatter(X,y)
plt.plot(X, lr.predict(X), color = 'red')
```
